# Data Exploration

```python
import os
import zipfile
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

## Unzip the output

```python
directory = 'cds_vs_pep'

for filename in os.listdir(directory):
    if os.path.isdir(directory+'/'+filename):
        print('folder')
    else:
        print(filename)
        if filename.split('.')[1] == 'zip':
            with zipfile.ZipFile(directory+'/'+filename, 'r') as zip_ref:
                zip_ref.extractall(directory+'/'+filename.split('.')[0])
```

```python
# Unzip the reproductive tissue
directory_repr = 'cds_repr_vs_pep'
for filename in os.listdir(directory_repr):
    if os.path.isdir(directory_repr+'/'+filename):
        print('folder')
    else:
        print(filename)
        if filename.split('.')[1] == 'zip':
            with zipfile.ZipFile(directory_repr+'/'+filename, 'r') as zip_ref:
                zip_ref.extractall(directory_repr+'/'+filename.split('.')[0])
```

## Import the summary files 

```python
directory = 'cds_vs_pep'

summary_list = []

for name in os.listdir(directory):
    if os.path.isdir(directory+'/'+name):
        print(name)
        with open(directory+'/'+name+'/fln_results/'+'summary_stats.txt') as f:
            lines = f.readlines()
            lines = np.array([ l.rstrip('\n').split('\t') for l in lines]).T
            df = pd.DataFrame(lines[np.newaxis,0], columns=lines[1])
            df['exp'] = name
            df['cds'] = name.split('_')[1]
            df['pep'] = name.split('_')[4]
            summary_list.append(df)
            
summary_df = pd.concat(summary_list)
summary_df
```

```python
for p in summary_df.pep.unique():
#sns.set_theme(style="whitegrid")
    print(p)
    melt_df = summary_df[summary_df['pep']==p].drop(columns=['exp','pep']).melt(id_vars=["cds"], var_name="variable") #.reset_index()
    melt_df = melt_df.astype({'value': 'float32'})
    g = sns.catplot(x="cds", 
                y="value", 
                kind="bar", 
                data=melt_df,
                col="variable", col_wrap=6,
                height=2,sharey=False,
                aspect=2, legend=True)

    for ax in g.axes:
        plt.setp(ax.get_xticklabels(), visible=True, rotation=45)

    plt.show()
```

## Import pt_seq

```python


pt_seq_list=[]

for name in os.listdir(directory):
    if os.path.isdir(directory+'/'+name):
        print(name)
        with open(directory+'/'+name+'/fln_results/'+'pt_seqs') as f:
            lines_raw = f.readlines()
            lines = []
            for l in lines_raw:
                l= l.rstrip('\n').split('\t')
                if len(l) == 25:
                    lines.append(l)
            lines = np.array(lines)
            #lines = np.array([ l.rstrip('\n').split('\t') for l in lines]) Changed because truncated db
            df = pd.DataFrame(lines[1:], columns=lines[0])
            df['exp'] = name
            df['cds'] = name.split('_')[1]
            df['pep'] = name.split('_')[4]   
            pt_seq_list.append(df)
            
pt_seq_df = pd.concat(pt_seq_list)
pt_seq_df
```

## Venns

```python
from venn import venn

sub_df = pt_seq_df[['Query_id', 'Subject_id','cds','pep']]
for p in summary_df.pep.unique():
    count_table = sub_df[sub_df['pep'] == p].drop(columns=['pep']).pivot_table(index='Subject_id', columns='cds', aggfunc=len, fill_value=0)
    reduced = count_table.drop(columns=[('Query_id',p)])
    reduced = reduced[(reduced.T != 0).any()]
    venn_d = {}
    for spe in list(reduced['Query_id'].columns):
        venn_d[spe] = set(list(reduced['Query_id'][spe][(reduced['Query_id'][spe] > 0)].index))
    print(p)
    venn(venn_d)
    plt.show()
```

## Hierachical clustering

```python
import seaborn as sns
from matplotlib.colors import LogNorm
    
sub_df = pt_seq_df[['Query_id', 'Subject_id','cds','pep']]
for p in summary_df.pep.unique():
    count_table = sub_df[sub_df['pep'] == p].drop(columns=['pep']).pivot_table(index='Subject_id', columns='cds', aggfunc=len, fill_value=0)
    reduced = count_table.drop(columns=[('Query_id',p)])
    reduced = reduced[(reduced.T != 0).any()]
    print(reduced.shape)
    print(p)
    sns.set_theme(color_codes=True)
    g = sns.clustermap(reduced, norm=LogNorm(), row_cluster=False)
    g.ax_heatmap.set(yticklabels=[])
    plt.show()
```
